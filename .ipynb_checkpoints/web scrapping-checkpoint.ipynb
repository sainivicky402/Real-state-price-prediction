```python
import requests
from bs4 import BeautifulSoup


def gets_latest_python_articles():
    url="https://www.python.org/"
    response=requests.get(url)

    if response.status_code==200:
        soup=BeautifulSoup(response.text,"html.parser")
        latest_articles=[]

        for article in soup.select(".blog-widget li"):
            title=article.a.text.strip()
            latest_articles.append(title)
            
        return latest_articles
    
    else:
        print(f"fail to retrieve information.status code:{response.status_code}")
        return []
    
if __name__=="__main__":
    python_articles=gets_latest_python_articles()

    if python_articles:
        print("New News in the python.org section")
        for index,article in enumerate(python_articles,1):
            print(f"{index}, {article}")
        
    else:
        print('no articles found')
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://facultytick.com/'
```

```python
response = requests.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    links = soup.find_all('a')
    for link in links:
        print(link.get('href'))
else:
    print(f"failed to retrieve the page. status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://quotes.toscrape.com"
```

```python
response = requests.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    quotes = []
    for quotes in soup.find_all('div', class_ = "Quotes"):
        text = quote.find('span', class_='text').get_text(strip=True)
        author = quote.find('small', class_='author').get_text(strip=True)
        quotes.append({'text':text, 'author':author})
    for quote in quotes:
        print(f'Text:{quote["text"]}\nAuthor: {quote["author"]}\n-----')

else:
    print(f'failed to retrieve.status code: {response.status_code}')
    
```
